## __1. Описание проекта__

**Задача:** Построение модели линейной регрессии для прогнозирования стоимости жилья в Калифорнии на основе данных 1990 года.

**Контекст:** Проект представляет собой классическую задачу регрессии, где необходимо предсказать медианную стоимость домов в жилых массивах на основе географических, демографических и экономических характеристик.

**Ключевые цели:**
1. Разработка точной модели для оценки стоимости недвижимости
2. Анализ факторов, наиболее сильно влияющих на цены на жильё
3. Оценка качества модели с помощью стандартных метрик регрессии

**Практическое применение:**
- Для **риелторов и агентств недвижимости** — автоматизация оценки объектов
- Для **инвесторов** — выявление перспективных районов для вложений
- Для **банков** — оценка залоговой стоимости при ипотечном кредитовании
- Для **муниципалитетов** — анализ рынка жилья и планирование развития районов

**Бизнес-ценность:**
- Снижение субъективности в оценке недвижимости
- Ускорение процесса оценки объектов
- Повышение точности прогнозов стоимости
- Возможность массовой оценки больших объёмов недвижимости


## __2. Описание данных__

**Название датасета:** California Housing Dataset (1990)

**Источник:** Перепись населения США 1990 года

**Объём данных:** ~20,000 записей (жилых массивов)

**Структура данных:**

| Признак | Тип данных | Описание | Единицы измерения |
|:---------|:------------|:----------|:-------------------|
| `longitude` | float64 | Долгота географического центра | градусы |
| `latitude` | float64 | Широта географического центра | градусы |
| `housing_median_age` | float64 | Медианный возраст зданий | годы |
| `total_rooms` | float64 | Общее количество комнат | штуки |
| `total_bedrooms` | float64 | Общее количество спален | штуки |
| `population` | float64 | Количество жителей | человек |
| `households` | float64 | Количество домовладений | единицы |
| `median_income` | float64 | Медианный доход жителей | десятки тыс. $ |
| `ocean_proximity` | object | Близость к океану | категория |
| `median_house_value` | float64 | Медианная стоимость дома | $ (Целевой признак) |

**Категории признака `ocean_proximity`:**
- `INLAND` — в глубине материка
- `NEAR OCEAN` — рядом с океаном
- `NEAR BAY` — рядом с заливом
- `ISLAND` — остров
- `<1H OCEAN` — менее 1 часа до океана

**Целевая переменная:**
- `median_house_value` — медианная стоимость дома в жилом массиве
- Диапазон значений: от $15,000 до $500,000
- Необходимо предсказать на основе остальных признаков

**Метрики оценки качества:**
1. **RMSE (Root Mean Square Error)** — среднеквадратичная ошибка
   - Измеряется в тех же единицах, что и целевая переменная ($)
   - Чувствительна к выбросам
   
2. **MAE (Mean Absolute Error)** — средняя абсолютная ошибка
   - Более устойчива к выбросам
   - Измеряется в долларах
   
3. **R² (Coefficient of Determination)** — коэффициент детерминации
   - Показывает долю объяснённой дисперсии
   - Диапазон: от 0 до 1 (чем ближе к 1, тем лучше)

**Ожидаемые сложности:**
1. Наличие выбросов в данных
2. Нелинейные зависимости между признаками
3. Влияние географического расположения на стоимость
4. Необходимость обработки категориальных признаков
5. Разные масштабы числовых признаков

## __5. Предобработка данных__

### 5.1 Замена пропусков

```python
df.describe().toPandas()
```

```python
display("Пропуски до обработки:")
for column in df.columns:
    null_count = df.filter(col(column).isNull()).count()
    print(f"{column}: {null_count} пропусков")
```

__Комментарий:__ Пропуск в колонке `total_bedrooms` допустимо заменить средним

```python
mean_value=df.select(mean('total_bedrooms')).collect()[0][0]
df=df.fillna(mean_value, subset='total_bedrooms')

display("Пропуски после обработки:")
for column in df.columns:
    null_count = df.filter(col(column).isNull()).count()
    print(f"{column}: {null_count} пропусков")
```

### 5.2 Преобразование категориальных колонок

__Комментарий:__ Распределим колонки по типам данных

```python
num_cols=['longitude','latitude','housing_median_age', 'total_rooms', 'total_bedrooms', 'total_rooms', 
     'households', 'median_income']
cat_cols=['ocean_proximity']
target='median_house_value'
```

```python
all_features=[num_cols, cat_cols]
```

__Комментарий:__ Поделим выборку на обучающую и тестовую, чтобы стандартизировать и масштабировать их отдельно

```python
train_df, test_df = df.randomSplit([0.8, 0.2], seed=RANDOM_SEED)
```

```python
indexer= StringIndexer(inputCols=cat_cols,
                      outputCols=[c+'_idx' for c in cat_cols],
                      handleInvalid="keep")

indexer_model=indexer.fit(train_df)

train_df=indexer_model.transform(train_df)

cols = [c for c in train_df.columns for i in cat_cols if (c.startswith(i))]

train_df.select(cols).show(3) 
```

```python
test_df=indexer_model.transform(test_df)

cols=[c for c in test_df.columns for i in cat_cols if (c.startswith(i))]

test_df.select(cols).show(3)
```

```python
encoder = OneHotEncoder(
    inputCols=[c + '_idx' for c in cat_cols],
    outputCols=[c + '_ohe' for c in cat_cols]
)

encoder_model=encoder.fit(train_df)
train_df=encoder_model.transform(train_df)
test_df=encoder_model.transform(test_df)
```

### 5.3 Трансформация числовых признаков

```python
assembler_num = VectorAssembler(
    inputCols=num_cols,
    outputCol="numerical_features"
)

train_df=assembler_num.transform(train_df)
test_df=assembler_num.transform(test_df)
```

```python
scaler = StandardScaler(
    inputCol="numerical_features",
    outputCol="scaled_numerical_features",
    withMean=True
)

scaler_model=scaler.fit(train_df)
train_df=scaler_model.transform(train_df)
test_df=scaler_model.transform(test_df)
```

### 5.4 Объединение признаков

```python
assembler_all = VectorAssembler(
    inputCols=['scaled_numerical_features', 'ocean_proximity_ohe'],  # ← ВАЖНО: используем МАСШТАБИРОВАННЫЕ признаки!
    outputCol="all_features"  
)

train_df = assembler_all.transform(train_df)
test_df = assembler_all.transform(test_df)
```

## __6. Обучение моделей и предсказания__

### 6.1 Модель 1(все данные)

__Коментарий:__ Обучим модель линейной регрессии на всех данных и сделаем предсказание

```python
lr_1=LinearRegression(labelCol=target, featuresCol="all_features")
model_1=lr_1.fit(train_df)

pred_1=model_1.transform(test_df)
predicted_labels_1=pred_1.select('median_house_value', 'prediction').withColumnRenamed("prediction", "pred_1")

predicted_labels_1.show()
```

### 6.2 Модель 2 (только числовые переменные)

__Комметарий:__ Отделим числовые признаки и заново обучим модель

```python
assembler_num = VectorAssembler(
    inputCols=['scaled_numerical_features'],  # ← ВАЖНО: используем МАСШТАБИРОВАННЫЕ признаки!
    outputCol="features_num"  
)

train_df = assembler_num.transform(train_df)
test_df = assembler_num.transform(test_df)

lr_2=LinearRegression(labelCol=target, featuresCol='features_num')
model_2=lr_2.fit(train_df)

pred_2=model_2.transform(test_df)
predicted_labels_2=pred_2.select('median_house_value', "prediction").withColumnRenamed("prediction", "pred_2")

predicted_labels_2.show(10)

```

### 6.3 Модель 3 (Random Forest)

__Комментарий:__ Удалим объединенные признаки и объединим их заново 

```python
train_df = train_df.drop("all_features")
test_df = test_df.drop("all_features")

train_df = assembler_all.transform(train_df)
test_df = assembler_all.transform(test_df)
```

```python
rf = RandomForestRegressor(
    featuresCol="all_features",      # столбец с признаками
    labelCol="median_house_value",  # целевая переменная
    predictionCol="prediction",  # столбец для предсказаний 
    numTrees=100,                # количество деревьев
    maxDepth=10,                 # максимальная глубина деревьев
    seed=42                      # seed для воспроизводимости
)

rf_model = rf.fit(train_df)

pred_3 = rf_model.transform(test_df)
predicted_labels_3=pred_3.select('median_house_value', "prediction").withColumnRenamed("prediction", "pred_3")

predicted_labels_3.show(10)
```

### 6.4 Анализ результатов
