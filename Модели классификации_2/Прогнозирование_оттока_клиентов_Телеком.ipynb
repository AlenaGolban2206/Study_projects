## __Описание проекта__

Оператор связи «ТелеДом» хочет бороться с оттоком клиентов. Для этого его сотрудники начнут предлагать промокоды и специальные условия всем, кто планирует отказаться от услуг связи. Чтобы заранее находить таких пользователей, «ТелеДому» нужна модель, которая будет предсказывать, разорвёт ли абонент договор. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и услугах. Ваша задача — обучить на этих данных модель для прогноза оттока клиентов.

## __Описание данных__
Данные хранятся в SQLite базе данных, состоящей из 4 взаимосвязанных таблиц:  
  

__1. Таблица <u>contract<u>__    
  
Поля таблицы:
    
      
`customerID` — ID абонента  
`BeginDate` — дата начала действия договора  
`EndDate` — дата окончания действия договора  
`Type` — тип оплаты: раз в год-два или ежемесячно  
`PaperlessBilling` — электронный расчётный лист  
`PaymentMethod` — тип платежа  
`MonthlyCharges` — расходы за месяц   
`TotalCharges` — общие расходы абонента  

Примечание: Информация о договорах актуальна на 1 февраля 2020.  
  
  

__2. Таблица <u>personal<u>__  

Поля таблицы:

`customerID` — ID абонента  
`gender` — пол  
`SeniorCitizen` — является ли абонент пенсионером  
`Partner` — есть ли у абонента супруг или супруга  
`Dependents` — есть ли у абонента дети    

  
__3. Таблица <u>internet<u>__  

Поля таблицы:

`customerID` — ID абонента  
`InternetService` — тип подключения  
`OnlineSecurity` — блокировка опасных сайтов  
`OnlineBackup` — облачное хранилище файлов для резервного копирования данных  
`DeviceProtection` — антивирус  
`TechSupport` — выделенная линия технической поддержки  
`StreamingTV` — стриминговое телевидение  
`StreamingMovies` — каталог фильмов  

  
__4. Таблица <u>phone<u>__

Поля таблицы:

`customerID` — ID абонента  
`MultipleLines` — подключение телефона к нескольким линиям одновременно    


___Связь между таблицами: Все таблицы связаны по полю <u>customerID<u>.___

```python
def cm_visual(y_test, y_pred, model_name=""):

    cm = confusion_matrix(y_test, y_pred)
    
    
    fig, ax = plt.subplots(figsize=(10, 10))
    
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                                  display_labels=['Не вкусно', 'Вкусно'])
    disp.plot(ax=ax, cmap='Blues', values_format='d')
    ax.set_title(f'Матрица ошибок - {model_name}', fontsize=14, pad=20)
    
    plt.tight_layout()
    plt.show()
    
    
    tn, fp, fn, tp = cm.ravel()
    
    print(f"True Negative (TN): {tn} - Правильно продлены")
    print(f"False Positive (FP): {fp} - Ложно расторгнуты") 
    print(f"False Negative (FN): {fn} - Пропущенные расторгнутые")
    print(f"True Positive (TP): {tp} - Правильно расторгнуты")
```

## __Подготовка данных для обучения__

__Комментарий:__ Разделим выборку на тренировочные и тестовые данные

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop(['customerID','Prolongation'], axis=1), 
                                                    data['Prolongation'], 
                                                    stratify=data['Prolongation'],
                                                    test_size=0.25,
                                                    random_state=RANDOM_STATE)
```

```python
copy=X_train.copy()

for column in copy:
    un=X_train[column].unique()
    print (f"Уникальные значения для признака:{column} - {len(un)}")
```

```python
ohe_columns=['PaperlessBilling','Partner', 'Dependents']
num_columns=['TotalCharges','Partnership_days','SeniorCitizen', 'Add_serv'] 

ord_columns=['Type', 'PaymentMethod', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',\
             'StreamingTV', 'StreamingMovies', 'MultipleLines']
```

__Комментарий:__ Создаем трансформеры для кодировки и масштабирования данных внутри пайплайна

```python
numeric_transformer = Pipeline(steps=[
    ('poly', PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)),
    ('scaler', StandardScaler())
])


data_preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, num_columns),
        ('cat', OneHotEncoder(handle_unknown='error', sparse_output=False), ohe_columns),
        ('ord', OrdinalEncoder(), ord_columns)
    ],
    remainder='passthrough'  
)

mlp_preprocessor = ColumnTransformer([
    ('ohe', OneHotEncoder(drop='first', sparse_output=False), ohe_columns),  
    ('num', StandardScaler(), num_columns),
    ('ordinal', OrdinalEncoder(), ord_columns)
], remainder='passthrough')

```

__Комментарий:__ Преобразовываем категориальные метки в числовые

```python
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)
```

## __Обучение модели__

__Комментарйи:__
    Для обучения создадим пайплайн

```python

models = [
    {
        'model': Pipeline([
            ('preprocessor', data_preprocessor),
            ('classifier', DecisionTreeClassifier(random_state=RANDOM_STATE, class_weight='balanced'))
        ]),
        'params': {
            'classifier__max_depth': list(range(5, 35, 5)) + [None],  
            'classifier__min_samples_split': list(range(2, 21, 2)),   
            'classifier__min_samples_leaf': list(range(1, 11, 1)),    
            'classifier__criterion': ['gini', 'entropy'],
            'classifier__max_features': ['sqrt', 'log2', None],
        }, 
        'name': 'DecisionTreeClassifier'
    },
    { 
    'model': Pipeline([
        ('preprocessor', data_preprocessor),
        ('classifier', RandomForestClassifier(
                random_state=RANDOM_STATE,
                n_jobs=1,
                class_weight='balanced'
            ))
        ]),
        'params':{
             
            'classifier__n_estimators': [1, 100, 600],
            'classifier__max_depth': [5, 10, 40, None],
            'classifier__min_samples_split': [2, 5, 10],
            'classifier__min_samples_leaf': [2, 5, 10],
            'classifier__max_features': ['sqrt', 'log2', None],
            'classifier__bootstrap': [True, False],
        },
        'name': 'RandomForestClassifier'
    },

    {
    'model': Pipeline([
        ('preprocessor', data_preprocessor),
        ('classifier', CatBoostClassifier(
            random_state=RANDOM_STATE,
            verbose=0,                    
            thread_count=1,               
            task_type='CPU',  
            auto_class_weights='Balanced'
           
        ))
    ]),
    'params': {
        'classifier__iterations': [100, 200, 300, 500],      
        'classifier__depth': [5, 7, 9, 10, 12],                   
        'classifier__learning_rate': [0.001, 0.01, 0.05, 0.1],
        'classifier__l2_leaf_reg': [1, 3, 5, 10],           
        'classifier__rsm': [0.7, 0.8, 0.9, 1.0],            
        'classifier__subsample': [0.7, 0.8, 0.9, 1.0],      
        
    }, 
    'name': 'CatBoostClassifier'
}, 
     {
        'model': Pipeline([
            ('preprocessor', mlp_preprocessor),
            ('classifier', MLPClassifier(
                random_state=RANDOM_STATE,
                max_iter=200,  
                verbose=0,
                solver='adam',
                early_stopping=True,  
                validation_fraction=0.1,
                n_iter_no_change=10,
                
            ))
        ]),
        'params': {
            'classifier__hidden_layer_sizes': [(50,), (100,), (5, ),  (6, 8)],  
            'classifier__activation': ['identity', 'logistic', 'tanh', 'relu'],  
            'classifier__alpha': [0.001, 0.01],  
            'classifier__learning_rate_init': [0.01, 0.002, 0.3, 0.005],  
            'classifier__batch_size': ['auto'],  
            'classifier__learning_rate': ['constant', 'adaptive']
        },
        'name': 'MLPClassifier'
     }
]
```

__Комментарйи студента:__ Теперь с помощью цикла выведем на экран лучшие результаты

```python
results = []

for model_vers in models:
    print('___________________________________________________')
    
    random_search = RandomizedSearchCV(
        estimator=model_vers['model'], 
        param_distributions=model_vers['params'],
        scoring='roc_auc',  
        n_jobs=1,
        cv=3,
        n_iter=20,  
        random_state=RANDOM_STATE,
        verbose=1   
    )

    
    random_search.fit(X_train, y_train_encoded)
    
    results.append({
        'name': model_vers['name'],
        'best_params': random_search.best_params_,
        'best_score': random_search.best_score_,
        'best_estimator': random_search.best_estimator_
    })
        
    print(f'Модель: {model_vers["name"]}')    
    print(f'Лучшие гиперпараметры: {random_search.best_params_}')
    print(f'Лучший ROC-AUC Score: {random_search.best_score_:.4f}')  # Изменено сообщение
    
    del random_search
    gc.collect()
    
```
