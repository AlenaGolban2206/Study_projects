## __Описание проекта__

Вы аналитик популярного сервиса аренды самокатов GoFast. Вам передали данные о некоторых пользователях из нескольких городов, а также об их поездках.   
  
Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. Сервисом можно пользоваться:

* без подписки
    - абонентская плата отсутствует;
    - стоимость одной минуты поездки — 8 рублей;
    - стоимость старта (начала поездки) — 50 рублей;

* с подпиской Ultra
    - абонентская плата — 199 рублей в месяц;
    - стоимость одной минуты поездки — 6 рублей;
    - стоимость старта — бесплатно.

<u>__Цель:__</u>   
    
Проанализировать данные и проверить гипотезы, которые могут помочь бизнесу вырасти.  



## __Описание данных__

 __1. Таблица <u>users_go.csv<u>__ 

 Поля таблицы:   

 `user_id` — уникальный идентификатор пользователя  
 `name` — имя пользователя  
 `age` — возраст  
 `city` — город  
 `subscription_type` — тип подписки (free, ultra)  

 __2. Таблица <u>rides_go.csv<u>__

 Поля таблицы:  

 `user_id` — уникальный идентификатор пользователя    
 `distance` — расстояние, которое пользователь проехал в текущей сессии (в метрах)  
 `duration` — продолжительность сессии (в минутах) — время с того момента, как пользователь нажал кнопку «Начать поездку» до момента, как он нажал кнопку «Завершить поездку»  
 `date` — дата совершения поездки  

 __3. Таблица <u>subscriptions_go.csv<u>__  

 Поля таблицы:  

 `subscription_type` — тип подписки  
 `minute_price` — стоимость одной минуты поездки по данной подписке  
 `start_ride_price` — стоимость начала поездки  
 `subscription_fee` — стоимость ежемесячного платежа

```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st
from scipy.stats import binom
from math import sqrt

# Настройки pandas
pd.options.display.float_format = '{:,.1f}'.format
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Библиотеки загружены")
```

## __Собственные функции__

```python
#Общий анализ данных

def total_analyse(data):
    
    def bold_text(text):
        return f"\033[1m{text}\033[0m"
    
    width = 100
    pd.set_option('display.width', width)
    
    print(f"\n{bold_text('Первые 5 строк:')}")
    print(data.head(5))
    print("-----------------")
    print(f"{bold_text('Последние 5 строк:')}")
    print(data.tail(5))
    print("------------------")
    print(f"{bold_text('Общая информация:')}")
    print(data.info())
    print("------------------") 
    print(f"{bold_text('Количество пропусков:')}")
    print(data.isna().sum())
    print("------------------") 
    print(f"{bold_text('Количество пустых строк:')}")
    print((data == '').sum().sum())
    print("------------------") 
    print(f"{bold_text('Количество дубликатов:')}")
    print(data.duplicated().sum())
```

```python
#Поиск уникальных значений

def unique_word(data):
    
    for column in data.columns:
        if data[column].dtype == 'object':
            
            print(f"\033[1mСтолбец\033[0m '{column}'")
            print(f"\033[1mУникальные значения:\033[0m {data[column].unique()}")
            print('------------------------------------------------------------------------------------')
            
```

```python
#Визуализация количественных и категориальных данных

def visual(data, column, title):
    
    if data[column].dtype in ['int64', 'float64', 'float32']:

    
        sns.set_theme(style='darkgrid')
    
        plt.figure(figsize=(15, 6))
        sns.histplot(data=data, x=column, kde=True, bins=40)
        plt.xlabel('Значение')
        plt.ylabel('Количество')
        plt.title(f'{title}', fontsize=16, fontweight='bold')
        plt.show()
    
        plt.figure(figsize=(15, 3))
        sns.boxplot(data=data, x=column, palette='viridis') 
        plt.xlabel('Значение')
        plt.title(f'Ящик с усами для колонки {column} ', fontsize=16, fontweight='bold')
        plt.show()
        
        print(f"\033[1mСтатистика:\033[0m")
        print(f" ")
        stats = data[column].describe()
        for metric, value in stats.items():
            value_str = f"{value:.1f}".replace(',', '')  
            print(f"{metric}: {value_str}")
        
    else:
        plt.figure(figsize=(15, 6))
        
        sns.countplot(data=data, x=column, palette='viridis')
        plt.title(f'{title}', fontsize=16, fontweight='bold')
        plt.xlabel('Значения', labelpad=20)
        plt.ylabel('Количество')

        plt.show()
```

#### Тестирование гипотезы 1

__Комментарий:__ Для этой задачи я собираюсь использовать гипотезу о равенстве средних двух генеральных совокупностей, так как я буду брать две генеральных выборки. Её смысл таков:

|Номер|Гипотеза|
|:--|:-----------|
|N0|C подпиской пользователи тратят времени на поездки столько же или меньше, чем без подписки|
|N1|Пользователи с подпиской времени на поездку тратят больше, чем без подписки|

```python
alpha = 0.05 #исторически сложившийся стандарт в статистике
results=st.ttest_ind(data_ultra_subscription['duration'], data_free_subscription['duration'], alternative = 'greater')

result_pvalue=results.pvalue
result_pvalue

display('p-значение:',result_pvalue)

if (result_pvalue<alpha) :
    display ('Нулевая гипотеза отвергнута, пользователи с подпиской времени на поездку тратят больше') 
else:
    display ('Не отвергаем нулевую гипотезу, с подпиской пользователи тратят времени столько же или меньше, чем без подписки')
```

|Номер|Гипотеза|
|:--|:-----------|
|N0|Пользователи с подпиской проезжают 3130 метров (или меньше)    
|N1|Пользователи с подпиской проезжают больше 3130 метров

__Комментарий:__ Тут я собираюсь проверить гипотезу о среднем двух генеральных выборок. Она будет выглядеть так: 

|Номер|Гипотеза|
|:--|:-----------|
|N0|Выручка одинаковая от обеих групп пользователей   
|N1|Выручка от пользователей с с подпиской выше, чем выручка от пользователей без подписки

## __Распределения__

### Акция с раздачей промокодов

Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. 
Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают
10 % пользователей. Необходимо выяснить, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %. 

__Комментарий:__ Необходимы параметры распределения, описывающего эту ситуацию и построение графика распределения.  
Используем биномиальное распределение, которое описывает указанную ситуацию. Затем подоберем подходящее значение параметра N для заданного параметра p=0.1 с помощью графиков и метода cdf().

```python
# Параметры задачи
p = 0.1               # Вероятность продления подписки
k = 100               # Минимальное количество успехов
target_prob = 0.05    # Целевая вероятность невыполнения плана

# Подбор минимального N
N = k                 # Начиная с минимального возможного значения
while True:
    distr = binom(N, p)        # Создаем объект биномиального распределения
    prob = distr.cdf(k - 1)    # P(X < k)
    if prob <= target_prob:
        break
    N += 1

print(f"Минимальное количество промокодов: {N}")

# Построение графика биномиального распределения для найденного N
x_values = np.arange(0, N + 1)
y_values = distr.pmf(x_values)    # Функция массы вероятности

plt.figure(figsize=(10, 6))
plt.bar(x_values, y_values, label=f"Binom(N={N}, p={p})", alpha=0.7)
plt.xlabel("Количество продлений (X)", fontsize=14)
plt.ylabel("Вероятность", fontsize=14)
plt.title(f"Биномиальное распределение количества продлений подписки\nN={N}, p={p}")
plt.axvline(x=k, color='red', linestyle='--', label=f'k = {k}')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

print(f"P(X < {k}) = {prob:.4f}")
```

__Комментарий:__
    
- Для достижения минимум 100 продлений с вероятностью продления p = 0.1 необходимо разослать 1161 промокод.
Это означает, что конверсия в продление составляет ~8.6% от общего количества промокодов, что близко к ожидаемой вероятности 10%.

  
- Вероятность получить менее 100 продлений при N = 1161 составляет ~4.98% (P(X < 100) ≈ 0.0498).
Это соответствует заданному порогу риска в 5%, что означает:

1. В 95% случаев план будет выполнен или перевыполнен.

2. Лишь в 5% случаев (примерно 1 из 20 кампаний) результат будет ниже планового.

### Push-уведомления в мобильном приложении

Отдел маркетинга рассылает клиентам push-уведомления в мобильном приложении. Клиенты могут открыть его или не открывать. Известно, что уведомления открывают около 40 % получивших клиентов. Отдел планирует разослать 1 млн уведомлений.

__Комментарий:__ С помощью аппроксимации необходимо построить примерный график распределения и оцените вероятность того, что уведомление откроют не более 
399,5 тыс. пользователей.

```python
from math import sqrt

n = 1000000
p = 0.4

# зададим мат.ожидание и ст.отклонение нормального распределения равными
# мат.ожиданию и ст.отклонению биномиального распределения
mu = n * p
sigma = sqrt(n * p * (1 - p))

# задаём нормальное распределение
distr = st.norm(mu, sigma) 

# считаем вероятность хвоста слева от 399,5 тысяч
result = distr.cdf(399500)
print(f"Вероятность того, что уведомление откроют не более 399,5 тыс. пользователеq: {result:.4f}%")

# Построение графика
x_values = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)  # Диапазон значений
y_values = distr.pdf(x_values)                                # Плотность вероятности

plt.figure(figsize=(10, 6))

plt.plot(x_values, y_values, label=f"N(μ={mu}, σ={sigma:.1f})")
plt.fill_between(x_values, y_values, where=(x_values <= 399500), color="skyblue", 
                 alpha=0.5, label=f"P(X ≤ 399500) = {result:.4f}")
plt.axvline(x=mu, color="red", linestyle="--", label=f"μ = {mu}")
plt.title("Распределение количества открытий уведомлений")
plt.xlabel("Количество открытий")
plt.ylabel("Плотность вероятности")
plt.legend()
plt.grid()

plt.show()
```

__Комментарий:__ 

- Вероятность получить ≤ 399,5 тыс. открытий: 15.37%
Это означает:

1. Примерно в каждой 6-7 рассылке результат будет ≤ 399,500
  
2. В ~84% случаев результат будет выше 399,5 тыс.

- Для увеличения вероятности достижения 400 тыс.+ можно:

1. Улучшить контент уведомлений (повысить p)

2. Увеличить выборку (n)

`5. Распределения`

[Акция с раздачей промокодов]

- Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают 10 % пользователей. 

_Цель:_

Необходимо выяснить, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно 5 %.

__Итоги:__

- Для достижения минимум 100 продлений с вероятностью продления p = 0.1 необходимо разослать 1161 промокод.
Это означает, что конверсия в продление составляет ~8.6% от общего количества промокодов, что близко к ожидаемой вероятности 10%.

  
- Вероятность получить менее 100 продлений при N = 1161 составляет ~4.98% (P(X < 100) ≈ 0.0498).
Это соответствует заданному порогу риска в 5%, что означает:

    1. В 95% случаев план будет выполнен или перевыполнен.

    2. Лишь в 5% случаев (примерно 1 из 20 кампаний) результат будет ниже планового.
