## __Описание проекта__

Я работаю в компании «Работа с заботой», которая специализируется на HR-аналитике и помогает организациям оптимизировать управление персоналом. Компания предоставила данные о сотрудниках, включая характеристики и результаты опросов об удовлетворённости работой.

В рамках проекта необходимо решить две ключевые задачи:

1. **Прогнозирование уровня удовлетворённости сотрудников** — построить модель, которая сможет предсказать показатель удовлетворённости (от 0 до 1) на основе доступных данных о сотрудниках. Это позволит сократить затраты на регулярные опросы и получить оперативную оценку настроений в коллективе.

2. **Прогнозирование оттока сотрудников** — создать модель, которая будет предсказывать вероятность увольнения сотрудников. Это критически важно для предотвращения внезапных потерь ключевых специалистов и снижения связанных с этим бизнес-рисков.

**Цель проекта:**  
Разработать систему машинного обучения, которая поможет компании:
- Снизить затраты на сбор данных через опросники
- Заблаговременно выявлять сотрудников с низкой удовлетворённостью работой
- Прогнозировать риски увольнения для принятия превентивных мер
- Оптимизировать процессы управления персоналом на основе данных

**Бизнес-ценность:**  
Проект позволит компании перейти от реактивного к проактивному управлению персоналом, снизить затраты на подбор и адаптацию новых сотрудников, повысить стабильность коллектива и улучшить общую эффективность HR-процессов.

## __Описание данных__

**<u>train_job_satisfaction_rate.csv<u>** - Тренировочная выборка данных о сотрудниках

**Признаки сотрудников:**
- `id` — уникальный идентификатор сотрудника
- `dept` — отдел, в котором работает сотрудник
- `level` — уровень занимаемой должности

- `salary` — ежемесячная зарплата сотрудника
- `job_satisfaction_rate` — уровень удовлетворённости сотрудника работой в компании (от 0 до 1) **Целевой признак**

```python
#Поиск уникальных значений

def unique_word(data):
    
    object_columns = data.select_dtypes(include=['object', 'category']).columns
        
    if len(object_columns) == 0:
        print("Нет категориальных колонок для обработки")
        return
        
    
        
    for column in object_columns:
            
        print(f"\033[1mСтолбец: '{column}'\033[0m")
        print(f"Количество уникальных значений: {data[column].nunique()}")
        print(f"\033[1mУникальные значения:\033[0m {data[column].unique()}")
        print('------------------------------------------------------------------------------------')
           
```

```python
#Визуализация количественных и категориальных данных

def visual(data, column, title):
    
    if data[column].dtype in ['int64', 'float64', 'float32'] and len(data[column].unique())>10:

    
        sns.set_theme(style='darkgrid')
    
        plt.figure(figsize=(15, 6))
        sns.histplot(data=data, x=column, kde=True, bins=40)
        plt.xlabel('Значение')
        plt.ylabel('Количество')
        plt.title(f'{title}', fontsize=16, fontweight='bold')
        plt.show()
    
        plt.figure(figsize=(15, 3))
        sns.boxplot(data=data, x=column, palette='viridis') 
        plt.xlabel('Значение')
        plt.title(f'Ящик с усами для колонки {column} ', fontsize=16, fontweight='bold')
        plt.show()
        
        print(f"\033[1mСтатистика:\033[0m")
        print(f" ")
        stats = data[column].describe()
        for metric, value in stats.items():
            value_str = f"{value:.1f}".replace(',', '')  
            print(f"{metric}: {value_str}")
        
    else:
        plt.figure(figsize=(15, 6))
        
        sns.countplot(data=data, x=column, palette='viridis')
        plt.title(f'{title}', fontsize=16, fontweight='bold')
        plt.xlabel('Значения', labelpad=20)
        plt.ylabel('Количество')

        plt.show()
```

```python
#Построение тепловой карты

def visual_corr(data, column_to_drop, interval_cols):
    
    plt.figure(figsize=(10,10))
    
    sns.heatmap((data.drop(column_to_drop, axis=1)).phik_matrix(interval_cols=interval_cols),annot=True, fmt='.2f', \
    cmap='viridis')
    
    
    plt.show()
```

```python
#Построение графиков «каждый с каждым»

def visual_pair(data, column_to_drop, hue):
    
    sns.pairplot(data.drop('Id', axis=1), hue='Порода', palette="viridis", corner=True, height=4)

    plt.show()
```

```python
def visual_compare(data_1, data_2, column_name_1, column_name_2, label_1, label_2, title, xlab, ylab):
    
    plt.figure(figsize=(10,10))
    
    sns.kdeplot(data=data_1, x=column_name_1, label=label_1, color='red', alpha=1, common_norm=False)
    sns.kdeplot(data=data_2, x=column_name_2, label=label_2, color='green', alpha=0.2, common_norm=False)
    
    plt.legend()
    plt.title(title)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    
    plt.show()
```

```python
def visual_perc(data, column):
    
    quited_all_new = (quited_all[column].value_counts(normalize=True) * 100).reset_index()
    quited_all_new.columns = [column, 'percentage']  

    
    plt.figure(figsize=(10, 6))
    sns.barplot(
        data=quited_all_new,
        x=column, 
        y='percentage',
        palette='viridis'  
    )

    
    plt.title('Количество увольнений по отделам', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('% увольнений', fontsize=12)
    plt.xticks(rotation=45)  
    plt.grid(axis='y', linestyle='--', alpha=0.7)  

    plt.show()
```

#### Собственная метрика

__Комментарий:__ 
    Метрика SMAPE вычисляется так:
Формула метрики SMAPE:

$$
\text{SMAPE} = \frac{100\%}{n} \sum_{i=1}^{n} \frac{|y_i - \hat{y}_i|}{(|y_i| + |\hat{y}_i|)/2}
$$

где:
- $y_i$ — фактическое значение целевого признака для объекта $i$
- $\hat{y}_i$ — предсказанное значение целевого признака для объекта $i$
- $n$ — количество объектов в выборке
- $\sum_{i=1}^{n}$ — сумма по всем объектам выборки

Нужно написать функцию, которая принимает на вход массивы NumPy или объекты Series в pandas и возвращает значение метрики SMAPE. 

```python
def smape(y_true, y_pred): 
    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)
    
    a = np.abs(y_true - y_pred)
    b = (np.abs(y_true) + np.abs(y_pred)) / 2
    
    # Избегаем деления на ноль (если и y_true, и y_pred равны нулю)
    mask = b == 0
    b[mask] = 1  # произвольное значение, так как b[mask] будет 0
    smape_values = a / b
    
    return 100 * np.mean(smape_values)
scorer = make_scorer(smape, greater_is_better=False)
```

#### Обучение Модели 1

```python
pipe_final= Pipeline([   
    ('preprocessor', data_preprocessor),
    ('models','passthrough')
])

param_grid = [ 
    
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
        'models__max_depth': range(2, 30),
        'models__min_samples_leaf': range(1, 10),
        'models__min_samples_split': range(2, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough'],
    },
    # словарь для модели LinearRegression()
    {
        'models': [LinearRegression()],
        'preprocessor__num': [StandardScaler(), 'passthrough'], 
        'preprocessor__ohe__ohe__drop': ['first']
    }
]



random_search_1 = RandomizedSearchCV(
    pipe_final,
    param_distributions=param_grid, 
    random_state=RANDOM_STATE,
    cv=5, 
    scoring=make_scorer(smape, greater_is_better=False),
    n_jobs=-1, 
    n_iter=20,
    error_score='raise'
)
    
random_search_1.fit(X_train, y_train) 
    
best_model_1 = random_search_1.best_estimator_
best_params_1 = random_search_1.best_params_


y_test_pred_1 = best_model_1.predict(X_test)
best_score_1=random_search_1.best_score_
test_smape_1 = smape(y_test, y_test_pred_1)



display("Лучшая модель:", best_model_1.named_steps['models'])
display("Параметры:", best_params_1)
display("Лучшая оценка", -best_score_1)
display(f"Test SMAPE: {test_smape_1:.4f}")
```

    - Подход позволяет использовать информацию из задачи регрессии (удовлетворённость) для задачи классификации (увольнение)

### __Подготовка данных__

```python
ohe_columns2=['dept','last_year_promo','last_year_violations']
ord_columns2=['level','workload']   
num_columns2=['employment_years','supervisor_evaluation','salary','DTR_predictions'] 

# создаём пайплайны для подготовки признаков
ohe_pipe2=Pipeline([('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore'))
                    ])

ord_pipe2=Pipeline([('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                   ('ord', OrdinalEncoder(categories=[['junior', 'middle', 'senior'], ['low','medium', 'high']],
                                          handle_unknown='use_encoded_value', unknown_value=np.nan)),
                   ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))])

num_pipe2=Pipeline([('simpleImputer_num', SimpleImputer(strategy='most_frequent')),
                   ('scaler', StandardScaler()),
                   ('poly', PolynomialFeatures(degree=2))])
                  

data_preprocessor2=ColumnTransformer([('ohe', ohe_pipe2, ohe_columns2),
                                  ('ord', ord_pipe2, ord_columns2),
                                  ('num', num_pipe2, num_columns2)],
                                   remainder='passthrough')
```

### __Обучение модели__

```python
pipe_final_2=Pipeline([
    
    ('preprocessor', data_preprocessor), 
    ('models', 'passthrough')
])

param_grid_2 = [
    # словарь для модели DecisionTreeClassifier()
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(2,20),
        'models__max_features': range(2,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']  
    },
    
    # словарь для модели KNeighborsClassifier() 
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2,10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']   
    },

    # словарь для модели LogisticRegression()
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, solver='saga'
        )],
        #'models__C':[0.1, 1, 10, 100],
        'models__C':[0.1, 1, 10],
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
        
    {
        'models':[SVC(probability=True,random_state=RANDOM_STATE)],
        'models__C':[0.1, 1, 10, 100]
    }]

cv = StratifiedKFold(n_splits=5, random_state=RANDOM_STATE, shuffle=True)

random_search_2 = RandomizedSearchCV(
    pipe_final_2,
    param_distributions=param_grid_2, 
    random_state=RANDOM_STATE,
    cv=cv,
    scoring='roc_auc',
    n_jobs=-1, 
    n_iter=50,
    error_score='raise'
)
    
random_search_2.fit(X_train_2, y_train_2) 
    
best_model_2 = random_search_2.best_estimator_
best_params_2 = random_search_2.best_params_

best_score_2 = random_search_2.best_score_
y_test_pred_2 = best_model_2.predict(X_test_2)
test_roc_auc = roc_auc_score(y_test_2, (best_model_2.predict_proba(X_test_2))[:,1])


display("Лучшая модель:", best_model_2.named_steps['models'])
display("Параметры:", best_params_2)
display(f"best roc_auc: {best_score_2:.4f}")
display(f"Test roc_auc: {test_roc_auc:.4f}")
 
```
