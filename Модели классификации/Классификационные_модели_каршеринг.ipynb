## __1. Описание проекта__

**Бизнес-контекст:**  
Вы — Data Scientist в каршеринговой компании. Компания хочет снизить риски аварийности и ущерба от ДТП, что позволит:
- **Уменьшить страховые выплаты**
- **Повысить безопасность клиентов**
- **Снизить затраты на ремонт автомобилей**
- **Улучшить репутацию компании**

**Проблема:**  
Сейчас водители бронируют и получают автомобили без какой-либо оценки рисков. Если водитель выбирает опасный маршрут или условия поездки рискованны, система не предупреждает его об этом.

**Цель проекта:**  
Разработать прототип системы оценки риска ДТП, которая:
1. **Предсказывает вероятность**, что конкретный водитель станет виновником ДТП
2. **Анализирует факторы риска** на основе исторических данных
3. **Даёт рекомендации** по безопасности  

## __2. Описание данных__

**<u>База данных data-science-vehicle-db</u>** — База данных о дорожно-транспортных происшествиях (ДТП)

**Таблицы в базе данных:**

---

**<u>collisions</u>** — Общая информация о ДТП

**Основные признаки:**
- `case_id` — уникальный идентификатор ДТП
- `collision_date` — дата происшествия
- `collision_time` — время происшествия
- `weather` — погодные условия
- `road_surface` — состояние дорожного покрытия
- `lighting` — освещение
- `road_condition` — общее состояние дороги
- `collision_severity` — тяжесть последствий ДТП
- `intersection` — произошло ли на перекрёстке
- `county` — округ
- `city` — город
- `state` — штат
- `location_type` — тип местности

---

**<u>parties</u>** — Информация об участниках ДТП

**Основные признаки:**
- `case_id` — идентификатор ДТП (связь с таблицей collisions)
- `party_number` — номер участника в рамках одного ДТП
- `party_type` — тип участника (например, "car" — машина)
- `at_fault` — является ли виновником (целевой признак для модели)
- `party_sobriety` — состояние трезвости участника
- `party_movement` — направление движения перед ДТП
- `party_age` — возраст участника
- `party_sex` — пол участника

---

**<u>vehicles</u>** — Информация о транспортных средствах

**Основные признаки:**
- `case_id` — идентификатор ДТП
- `party_number` — номер участника
- `vehicle_make` — марка автомобиля
- `vehicle_model` — модель автомобиля
- `vehicle_year` — год выпуска автомобиля
- `vehicle_age` — возраст автомобиля (в годах) **Важный фактор по требованию заказчика**
- `vehicle_damage` — степень повреждения транспортного средства
- `damage_severity` — серьёзность повреждений

---

**<u>Связь между таблицами</u>**
- Все таблицы связаны через `case_id`
- Для уникальной идентификации участника ДТП используется комбинация `case_id` + `party_number`
- Таблица `collisions` содержит общую информацию о происшествии
- Таблица `parties` содержит данные о каждом участнике ДТП
- Таблица `vehicles` содержит технические данные о транспортных средствах

---

**<u>Основные ограничения для модели</u>**
1. **Тип участника**: только автомобили (`party_type = 'car'`)
2. **Серьёзность повреждений**: все повреждения, кроме царапин (`damage_severity != 'SCRATCH'`)
3. **Период данных**: только 2012 год
4. **Обязательный фактор**: возраст автомобиля должен быть учтён
5. **Целевая переменная**: `at_fault` (является ли участник виновником ДТП)

---

**<u>Дополнительная информация</u>**
- Все таблицы имеют данные за период наблюдений
- Общий ключ `case_id` позволяет связать все таблицы
- Для анализа доступны как временные, так и пространственные характеристики ДТП
- Имеется возможность анализа влияния погодных условий, состояния дороги и других факторов на вероятность ДТП

```python
def visual_cat(data, column):
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    unique_count = data[column].nunique()
    value_counts = data[column].value_counts()
    
    
    top_values = value_counts.head(10)
    bars = ax1.bar(range(len(top_values)), top_values.values)
    ax1.set_xticks(range(len(top_values)))
    ax1.set_xticklabels(top_values.index, rotation=45, ha='right')
    ax1.set_title(f'Bar Plot: {column}')
    ax1.set_xlabel('Значения')
    ax1.set_ylabel('Количество')
    
    
    value_counts_pie = value_counts.head(5)
    ax2.pie(value_counts_pie.values, labels=value_counts_pie.index, autopct='%1.1f%%')
    ax2.set_title(f'Pie Chart: {column}')
    ax2.axis('equal')
    
    plt.tight_layout()
    plt.show()
    
    print(f"Уникальных значений: {unique_count}")
```

```python
def scatter_plot(columns_pairs, special_feat, data):
    
    fig, axes = plt.subplots(2, 3, figsize=(25, 15))
    axes = axes.ravel()

    n_pairs = len(columns_pairs)
    
    for i, (x_feat, y_feat) in enumerate(columns_pairs):
        scatter = axes[i].scatter(
            x=data[x_feat], 
            y=data[y_feat],
            c=data[special_feat],
            cmap='viridis', 
            s=40,
            alpha=0.7
        )
        axes[i].set_xlabel(x_feat, fontsize=12)
        axes[i].set_ylabel(y_feat, fontsize=12)
        axes[i].set_title(f'{x_feat} vs {y_feat}', fontsize=14)
        axes[i].tick_params(axis='both', which='major', labelsize=11)
    
        legend = plt.colorbar(scatter, ax=axes[i])
        legend.set_label(special_feat, fontsize=12)
       
    for i in range(n_pairs, len(axes)):
        axes[i].set_visible(False)
    
    plt.tight_layout()
    plt.show() 
    
    #функция для построения диаграммы рассеяния по признаку
```

__Комментарий:__ Среди моделей были выбраны следующие:

- CatBoost - так как работает с категориальными,  устойчивость к переобучению

```python
models = [
    {
        'model': CatBoostClassifier(random_state=RANDOM_STATE, verbose=False),
        'params': {
            'iterations': [200,  400, 800, 1000],
            'depth': [2, 4, 6, 8, 10, 20],
            'learning_rate': [0.01, 0.05, 0.07, 0.1, 0.2],
        }, 
        'name': 'CatBoostClassifier'
    },

    {
        'model': LogisticRegression(random_state=RANDOM_STATE),
        'params': {
            'C': [0.1, 1.0, 10.0],
            'penalty': ['l1', 'l2'],
            'solver': ['liblinear'],
            'max_iter': [1000],
            'class_weight': ['balanced', None],
        },
        'name': 'LogisticRegression'
    },
    
    {
        'model': KNeighborsClassifier(n_jobs=-1),
        'params': {
            'n_neighbors': [3, 6],
            'weights': ['uniform'],
            'metric': ['euclidean'],
            'leaf_size': [20, 30],
        },
        'name': 'KNeighborsClassifier'
    }
]
```

### _9.2. Выбор метрики для оценки модели, исходя из поставленной бизнесом задачи._

__Комментарий:__ В данной задаче выбор был сделан на метрику `f1`.  

Задача бизнеса - определить вероятность ДТП - 1 или 0. У нас нет понимания, что важнее - False Positive (ДТП не случилось, а модель выдала наоборот), или False Negative (ДТП случилось, а модель выдала наоборот), поэтому нам важно найти баланс между precision и recall, то есть метрику 'f1'.

```python
warnings.filterwarnings('ignore')

results = []

for model_vers in models:
    print('___________________________________________________')
    
    random_search = RandomizedSearchCV(
        estimator=model_vers['model'], 
        param_distributions=model_vers['params'],
        scoring='f1',
        n_jobs=1,
        cv=3,
        n_iter=20,  
        random_state=RANDOM_STATE,
        verbose=1   
    )

    random_search.fit(X_train, y_train)
    
    results.append({
        'name': model_vers['name'],
        'best_params': random_search.best_params_,
        'best_score': random_search.best_score_,
        'best_estimator': random_search.best_estimator_
    })
        
    print(f'Модель: {model_vers["name"]}')    
    print(f'Лучшие гиперпараметры: {random_search.best_params_}')
    print(f'Лучший F1 Score: {random_search.best_score_:.4f}')    
    
    del random_search
    gc.collect()
```


### _9.3. Оформите вывод в виде сравнительной таблицы._

```python
results_df = pd.DataFrame(results)
best_overall = results_df.loc[results_df['best_score'].idxmax()]         
                              
results_df[['name', 'best_params', 'best_score']]                              
```

### 10.3. _Максимизация Recall_

```python
y_proba = best_model.predict_proba(X_test)[:,1]
data_new = pd.DataFrame(zip(y_test, y_proba),
             columns = ['y_valid', 'y_proba']).sort_values(by='y_proba',ascending=False)

thresholds = [round(i,2) for i in np.linspace(0.1,1,num = 4,endpoint=False)]

columns = []

for i in thresholds:
    columns.append('y_pred_'+str(i))
    data_new['y_pred_'+str(i)] = data_new['y_proba'].apply(lambda x: 1 if x>=i else 0) 
     

    data_new.sample(5)
    
y_true = data_new['y_valid']


results = []

for i in thresholds:
    y_pred = data_new['y_pred_' + str(i)]
    recall = recall_score(y_true, y_pred)
    predicted_count = y_pred.sum()
    total_class_1 = y_true.sum()

    results.append([round(i,2), recall, predicted_count, total_class_1])


results_df = pd.DataFrame(results, columns=['Порог', 'recall', 'Число предсказанных объектов', 'Всего объектов класса 1'])
print(results_df)
```
